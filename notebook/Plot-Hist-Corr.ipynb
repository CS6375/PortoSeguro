```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns
```

```python
train = pd.read_csv("../data/train.csv", na_values=['-1'])
```

```python
train.describe()
```

```python
no_id_train = train.drop(['id'], axis=1)
```

```python
cat_col = list(filter(lambda x : 'cat' in x, [x for x in train.columns]))
bin_col = list(filter(lambda x : 'bin' in x, [x for x in train.columns]))
num_col = list(filter(lambda x : not ('id' in x or 'bin' in x or 'cat' in x), [x for x in train.columns]))

ind_col = list(filter(lambda x : 'ind' in x, [x for x in train.columns]))
reg_col = list(filter(lambda x : 'reg' in x, [x for x in train.columns]))
car_col = list(filter(lambda x : 'car' in x, [x for x in train.columns]))
calc_col = list(filter(lambda x : 'calc' in x, [x for x in train.columns]))
```

```python
plt.figure()
no_id_train[ind_col].hist(histtype='bar', 
                          normed=True, 
                          color='k', 
                          alpha=0.6, 
                          bins=50, 
                          layout=(3, 6), 
                          figsize = (24, 7.5))
plt.savefig('ind_col.pdf', bbox_inches='tight')
plt.close()
```

```python
plt.figure()
no_id_train[reg_col].hist(histtype='bar', 
                          normed=True, 
                          color='k', 
                          alpha=0.6, 
                          bins=50, 
                          layout=(1, 3), 
                          figsize = (12, 2.5))
plt.savefig('reg_col.pdf', bbox_inches='tight')
plt.close('all')
```

```python
plt.figure()
no_id_train[car_col].hist(histtype='bar', 
                          normed=True, 
                          color='k', 
                          alpha=0.6, 
                          bins=50, 
                          layout=(3, 6), 
                          figsize = (24, 7.5))
plt.savefig('car_col.pdf', bbox_inches='tight')
plt.close('all')
```

```python
plt.figure()
no_id_train[calc_col].hist(histtype='bar', 
                          normed=True, 
                          color='k', 
                          alpha=0.6, 
                          bins=50, 
                          layout=(4, 6), 
                          figsize = (24, 10))
plt.savefig('calc_col.pdf', bbox_inches='tight')
plt.close('all')
```

```python
sns.set(style="white")

plt.figure()

# Compute the correlation matrix
corr = train[ind_col].corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=False, linewidths=.5, cbar_kws={"shrink": .5})

plt.savefig('corr_ind_col.pdf', bbox_inches='tight')
plt.close('all')
```
